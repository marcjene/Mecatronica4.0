**Paso 7: Normalización de los datos** (ej: usando el MinMaxScaler)

`(0.5)`

```python
# Hacer una normalización de todos los atributos (Sin considerar 'is_safe')
data_features = data.iloc[:, :-1]

# Importar MinMaxScaler
from sklearn.preprocessing import MinMaxScaler

# Escalar los datos
scaler = MinMaxScaler()
data_features_sc = data_features.copy()
data_features_sc = pd.DataFrame(scaler.fit_transform(data_features))
data_features_sc.columns = data_features.columns
data_features_sc.head()
```

**Paso 8: Dividir los datos**

En este paso se tendra que realizar la división de los datos para que quede aproximadamente:
- Datos de entrenamiento 70%
- Datos de Validación 15%
- Datos de Prueba 15%.

Primero, divido los datos en:

- Features/características/atributos/variables de entrada: X
- Target/etiquedas: Y

`(0.3 por dividir correctamente X e Y)`

`(0.3 por dividir correctamente train, test, validation)`

```python
# Atributos X (features); etiquetas y (target) y transformar a numpy
X = data_features_sc
X = X.to_numpy()

Y = data[["is_safe"]]
Y = Y.to_numpy()
```

```python
# Importar train_test_split
from sklearn.model_selection import train_test_split

# Fijar un porcentaje de "test_size"
test_size = 0.15  # porcentaje de los datos de entrada que utilizaré para validar el modelo

# Divido los datos en datos de entreno, validación y prueba

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=test_size, shuffle=False)

X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=test_size, shuffle=False)
```

**Paso 9: Construcción y evaluación de modelos**

Usar (como mínimo) los siguientes modelos:

- Logistic Regression
- Random Forest Classifier
- KNeighbors Classifier

Responder a las siguientes preguntas:
* **¿Cuál obtiene mejores resultados?** 
* **¿Qué balanced_accuracy obtiene?** 

`(0.5: 0.4 por hacer la comparación, 0.1 por responder correctamente las preguntas)`

```python
# Importar modelos
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
# Definir folds, metrica de error (ej: balanced_accuracy) y crear modelos
num_folds = 5
error_metrics = {'balanced_accuracy'}
models = { ('LR', LogisticRegression()), 
          ('KNN', KNeighborsClassifier()),
           ('RF', RandomForestClassifier())
         }

results = [] # guarda los resultados de las métricas de evaluación
names = []  # Nombre de cada algoritmo
msg = []  # imprime el resumen del método de cross-validation
```

```python
# Importar librerias necesarias
from sklearn.model_selection import cross_val_score, GridSearchCV
from sklearn.model_selection import StratifiedKFold, TimeSeriesSplit

# Entrenar con validación cruzada y comparar resultados entre algoritmos
for scoring in error_metrics:
    print('Métrica de evaluación: ', scoring)
    for name, model in models:
        print('Modelo ', name)
        cross_validation = TimeSeriesSplit(n_splits=num_folds)
        cv_results = cross_val_score(model, X_train, np.ravel(y_train), cv=cross_validation, scoring=scoring)
        results.append(cv_results)
        names.append(name)
        resume = (name, cv_results.mean(), cv_results.std())
        msg.append(resume)
    print(msg)

    # Comparar resultados entre algoritmos
    sns.set(rc={'figure.figsize':(8,5)})
    fig = plt.figure()
    fig.suptitle('Comparación de algoritmos con métrica de evaluación: %s' %scoring)
    ax = fig.add_subplot(111)
    ax.set_xlabel('Modelos candidatos')
    ax.set_ylabel('%s' %scoring)
    plt.boxplot(results)
    ax.set_xticklabels(names)
    plt.show()

    results = []
```

Paso 10: Ajustar hiperparámetros

Pasos para realizar el hiperajuste de los parámetros:

```python
# Importar librerias necesarias
from sklearn.model_selection import TimeSeriesSplit

# Definir hiperparámetros para el modelo con mejores resultados
modelo = RandomForestClassifier()
hyperparams = {
     'n_estimators': [100, 500, 800], #default=100
     'min_samples_split': [2, 4], #default=2
     'bootstrap': [True, False],  #default=True
 }

# Hacer validación cruzada
scoring='balanced_accuracy'
cross_validation = TimeSeriesSplit(n_splits=10)
my_cv = cross_validation.split(X_val, y_val)
gsearch = GridSearchCV(estimator=modelo, param_grid=hyperparams, scoring=scoring, cv=my_cv, verbose=2) 
gsearch.fit(X_val, np.ravel(y_val))

print("Mejor resultado: %f utilizando los siguientes hiperparámetros %s" % (gsearch.best_score_, gsearch.best_params_))
means = gsearch.cv_results_['mean_test_score']
stds = gsearch.cv_results_['std_test_score']
params = gsearch.cv_results_['params']
```

**Paso 11: Evaluación final del modelo**

Métricas de evaluación:
  * 1. Matriz de confusión
  * 2. Coeficiente de Matthews (MCC)
  * 3. Curva ROC/AUC

    
Entrena el modelo con los hiperparámetros óptimos encontrados en el apartado anterior y realiza las predicciones. 

`(0.6: 0.2 por métrica)`

```python
# Definir modelo final con hiperparámetros óptimos
modelo_final =RandomForestClassifier(bootstrap=False, min_samples_split=4, n_estimators= 100)
modelo_final.fit(X_train, np.ravel(y_train))

# Realizar las predicciones con X_test
y_predict = modelo_final.predict(X_test)
```

```python
# Importar librerías necesarias
from sklearn.metrics import confusion_matrix, classification_report, plot_confusion_matrix

# Matriz de confusión
confusion_matrix = confusion_matrix(y_test, y_predict)
print(classification_report(y_test, y_predict))
print(confusion_matrix)
```

```python
# Importar libreria y calcular MCC
from sklearn.metrics import matthews_corrcoef

matthews_corrcoef(y_test, y_predict)
```

```python
# Curva ROC/AUC
from sklearn.metrics import roc_auc_score, plot_roc_curve, roc_curve, auc

plot_roc_curve(modelo_final, X_test, y_test)
auc = roc_auc_score(y_test, y_predict)
```
